## Features

### Administrative

```python
fig = px.histogram(online_shoppers, x="Administrative", title="Distribution of the Number of Administrative Page Views")

fig.update_layout(
    legend_title_text = "Administrative",
    yaxis_title = "Frequency",
    xaxis_title = "Administrative Page Views"
)

fig.show()
```
